(a) Create two random variables ...


- Call them "vec_x" and "vec_z"
- Generate "vec_x" from a uniform distribution with (low = 10, high = 20, size = 1000)
- Generate "vec_z" from a uniform distribution with (low = -2, high = 5, size = 1000)
- Plot a histogram of each variable
    - Label the axes and title
    - Each graph should have a different title
 
To get full points, plot the histograms on a grid (See Lecture 6)

````python
# Initialize and assign values to vec_x and vec_z
vec_x = np.random.uniform(low = 10, high = 20, size = 1000)
vec_z = np.random.uniform(low = -2, high = 5, size = 1000)

#create iterable list of vectors
vectors = [vec_x,vec_z]
#create iterable list of vector names
vectorNames = ["Vector X","Vector Z"]

#subplot creation
fig, sub_list = plt.subplots(1,2,figsize=(8,4))
plt.tight_layout() # to prevent overlap

# create histograms using for loop
ind = 0
for v in vectors:
    # histogram for vector
    sub_list[ind].hist(x=v)
    sub_list[ind].set_title(vectorNames[ind]+"'s Uniform Distribution")
    sub_list[ind].set_xlabel("Values")
    sub_list[ind].set_ylabel("Frequency")
    ind = ind + 1
````

- Initiate a for loop over the list of dataset names. Inside the loop:

Note: Check that you did it correctly by trying to open your new excel files outside "VSCode"

````python
# creating lists with names
dataset_fullnames_csv = []
list_newnames = []
subfolder_csv = "data_quiz/"
subfolder_xlsx = "data_excel/"
for name in dataset_names:
    dataset_fullnames_csv.append(subfolder_csv + name + ".csv")
    list_newnames.append(subfolder_xlsx + name + ".xlsx")

# start index
index = 0

# change csv to xlsx
for fullname in dataset_fullnames_csv:
    dataset = pd.read_csv(fullname)
    filename = list_newnames[index]
    # store dataset as an excel file with name filename
    dataset.to_excel(filename)

    index = index + 1
````
